```python
def funct(error):
    file_name = []
    i = 0
    N_val_bourse = []
    N_val_climat = []
    N_val_dd = []
    N_val_energies = []
    N_val_maladies = []
    
    S_val_bourse = []
    S_val_climat = []
    S_val_dd = []
    S_val_energies = []
    S_val_maladies = []
    
    P_val_bourse = []
    P_val_climat = []
    P_val_dd = []
    P_val_energies = []
    P_val_maladies = []
    
    T_val_bourse = []
    T_val_climat = []
    T_val_dd = []
    T_val_energies = []
    T_val_maladies = []
    
    h = np.array([10, 100, 200, 300])
    for files in glob.glob("*.csv"):
        file_name.append(Path(files).stem)
        #print(file_name[i])
        df = pd.read_csv(files)
        if(file_name[i][0] == '1'):
            if(file_name[i][9] == 'M'):
                N_val_bourse.append(df[error][0])
                N_val_climat.append(df[error][1])
                N_val_dd.append(df[error][2])
                N_val_energies.append(df[error][3])
                N_val_maladies.append(df[error][4])
            elif(file_name[i][9] == 'P'):
                P_val_bourse.append(df[error][0])
                P_val_climat.append(df[error][1])
                P_val_dd.append(df[error][2])
                P_val_energies.append(df[error][3])
                P_val_maladies.append(df[error][4])
            elif(file_name[i][9] == 'S'):
                S_val_bourse.append(df[error][0])
                S_val_climat.append(df[error][1])
                S_val_dd.append(df[error][2])
                S_val_energies.append(df[error][3])
                S_val_maladies.append(df[error][4])
            elif(file_name[i][9] == 'T'):
                T_val_bourse.append(df[error][0])
                T_val_climat.append(df[error][1])
                T_val_dd.append(df[error][2])
                T_val_energies.append(df[error][3])
                T_val_maladies.append(df[error][4])
        elif(file_name[i][5] == '1'):
            if(file_name[i][9] == 'M'):
                N_val_bourse.append(df[error][0])
                N_val_climat.append(df[error][1])
                N_val_dd.append(df[error][2])
                N_val_energies.append(df[error][3])
                N_val_maladies.append(df[error][4])
                
            elif(file_name[i][9] == 'P'):
                P_val_bourse.append(df[error][0])
                P_val_climat.append(df[error][1])
                P_val_dd.append(df[error][2])
                P_val_energies.append(df[error][3])
                P_val_maladies.append(df[error][4])
            elif(file_name[i][9] == 'S'):
                S_val_bourse.append(df[error][0])
                S_val_climat.append(df[error][1])
                S_val_dd.append(df[error][2])
                S_val_energies.append(df[error][3])
                S_val_maladies.append(df[error][4])
            elif(file_name[i][9] == 'T'):
                T_val_bourse.append(df[error][0])
                T_val_climat.append(df[error][1])
                T_val_dd.append(df[error][2])
                T_val_energies.append(df[error][3])
                T_val_maladies.append(df[error][4])
        elif(file_name[i][5] == '2'):
            if(file_name[i][9] == 'M'):
                N_val_bourse.append(df[error][0])
                N_val_climat.append(df[error][1])
                N_val_dd.append(df[error][2])
                N_val_energies.append(df[error][3])
                N_val_maladies.append(df[error][4])
                
            elif(file_name[i][9] == 'P'):
                P_val_bourse.append(df[error][0])
                P_val_climat.append(df[error][1])
                P_val_dd.append(df[error][2])
                P_val_energies.append(df[error][3])
                P_val_maladies.append(df[error][4])
            elif(file_name[i][9] == 'S'):
                S_val_bourse.append(df[error][0])
                S_val_climat.append(df[error][1])
                S_val_dd.append(df[error][2])
                S_val_energies.append(df[error][3])
                S_val_maladies.append(df[error][4])
            elif(file_name[i][9] == 'T'):
                T_val_bourse.append(df[error][0])
                T_val_climat.append(df[error][1])
                T_val_dd.append(df[error][2])
                T_val_energies.append(df[error][3])
                T_val_maladies.append(df[error][4])
        elif(file_name[i][5] == '3'):
            if(file_name[i][9] == 'M'):
                N_val_bourse.append(df[error][0])
                N_val_climat.append(df[error][1])
                N_val_dd.append(df[error][2])
                N_val_energies.append(df[error][3])
                N_val_maladies.append(df[error][4])
                
            elif(file_name[i][9] == 'P'):
                P_val_bourse.append(df[error][0])
                P_val_climat.append(df[error][1])
                P_val_dd.append(df[error][2])
                P_val_energies.append(df[error][3])
                P_val_maladies.append(df[error][4])
            elif(file_name[i][9] == 'S'):
                S_val_bourse.append(df[error][0])
                S_val_climat.append(df[error][1])
                S_val_dd.append(df[error][2])
                S_val_energies.append(df[error][3])
                S_val_maladies.append(df[error][4])
            elif(file_name[i][9] == 'T'):
                T_val_bourse.append(df[error][0])
                T_val_climat.append(df[error][1])
                T_val_dd.append(df[error][2])
                T_val_energies.append(df[error][3])
                T_val_maladies.append(df[error][4])
        i = i +1
        
    plt.figure(figsize=(10,5))
    plt.plot(h, N_val_bourse, label = 'Bourse', marker="+")
    plt.plot(h, N_val_climat, label = 'Climat', marker="+")
    plt.plot(h, N_val_dd, label = 'Données démographiques', marker="+")
    plt.plot(h, N_val_energies, label = 'Energies', marker="+")
    plt.plot(h, N_val_maladies, label = 'Maladies', marker="+")
    plt.legend()
    plt.title("Evolution de la "+ error + " en fonction de l'horizon de prédiction (Méthode naïve)",fontsize=10)
    plt.show()
    
    plt.figure(figsize=(10,5))
    plt.plot(h, S_val_bourse, label = 'Bourse', marker="+")
    plt.plot(h, S_val_climat, label = 'Climat', marker="+")
    plt.plot(h, S_val_dd, label = 'Données démographiques', marker="+")
    plt.plot(h, S_val_energies, label = 'Energies', marker="+")
    plt.plot(h, S_val_maladies, label = 'Maladies', marker="+")
    plt.legend()
    plt.title("Evolution de la "+ error + " en fonction de l'horizon de prédiction (Sarima)",fontsize=10)
    plt.show()
    
    plt.figure(figsize=(10,5))
    plt.plot(h, P_val_bourse, label = 'Bourse', marker="+")
    plt.plot(h, P_val_climat, label = 'Climat', marker="+")
    plt.plot(h, P_val_dd, label = 'Données démographiques', marker="+")
    plt.plot(h, P_val_energies, label = 'Energies', marker="+")
    plt.plot(h, P_val_maladies, label = 'Maladies', marker="+")
    plt.legend()
    plt.title("Evolution de la "+ error + " en fonction de l'horizon de prédiction (Prophet)",fontsize=10)
    plt.show()
    
    plt.figure(figsize=(10,5))
    plt.plot(h, T_val_bourse, label = 'Bourse', marker="+")
    plt.plot(h, T_val_climat, label = 'Climat', marker="+")
    plt.plot(h, T_val_dd, label = 'Données démographiques', marker="+")
    plt.plot(h, T_val_energies, label = 'Energies', marker="+")
    plt.plot(h, T_val_maladies, label = 'Maladies', marker="+")
    plt.legend()
    plt.title("Evolution de la "+ error + " en fonction de l'horizon de prédiction (Transformer)",fontsize=10)
    plt.show()
```

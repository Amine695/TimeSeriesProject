```python
file_name = []
resultats_prophet = dict.fromkeys(["Bourse", "Climat", "Données démographiques", "Energies", "Maladies"])
resultats_sarima = dict.fromkeys(["Bourse", "Climat", "Données démographiques", "Energies", "Maladies"])
resultats_naive = dict.fromkeys(["Bourse", "Climat", "Données démographiques", "Energies", "Maladies"])
resultats_transformers = dict.fromkeys(["Bourse", "Climat", "Données démographiques", "Energies", "Maladies"])
i=0

for files in glob.glob("*.csv"):
    file_name.append(Path(files).stem)
    print(file_name[i])
    df = pd.read_csv(files)
    if(file_name[i][0] == 'p'):
        if(file_name[i][16] == 'b'):
            resultats_prophet["Bourse"] = df
        elif(file_name[i][16] == 'c'):
            resultats_prophet["Climat"] = df
        elif(file_name[i][16] == 'd'):
            resultats_prophet["Données démographiques"] = df
        elif(file_name[i][16] == 'e'):
            resultats_prophet["Energies"] = df
        elif(file_name[i][16] == 'm'):
            resultats_prophet["Maladies"] = df
    if(file_name[i][0] == 'S'):
        if(file_name[i][15] == 'b'):
            resultats_sarima["Bourse"] = df
        elif(file_name[i][15] == 'c'):
            resultats_sarima["Climat"] = df
        elif(file_name[i][15] == 'd'):
            resultats_sarima["Données démographiques"] = df
        elif(file_name[i][15] == 'e'):
            resultats_sarima["Energies"] = df
        elif(file_name[i][15] == 'm'):
            resultats_sarima["Maladies"] = df
    if(file_name[i][0] == 'n'):
        if(file_name[i][14] == 'b'):
            resultats_naive["Bourse"] = df
        elif(file_name[i][14] == 'c'):
            resultats_naive["Climat"] = df
        elif(file_name[i][14] == 'd'):
            resultats_naive["Données démographiques"] = df
        elif(file_name[i][14] == 'e'):
            resultats_naive["Energies"] = df
        elif(file_name[i][14] == 'm'):
            resultats_naive["Maladies"] = df
    if(file_name[i][0] == 'T'):
        if(file_name[i][20] == 'b'):
            resultats_transformers["Bourse"] = df
        elif(file_name[i][20] == 'c'):
            resultats_transformers["Climat"] = df
        elif(file_name[i][20] == 'd'):
            resultats_transformers["Données démographiques"] = df
        elif(file_name[i][20] == 'e'):
            resultats_transformers["Energies"] = df
        elif(file_name[i][20] == 'm'):
            resultats_transformers["Maladies"] = df
    i = i+1
```

```python
s = ["MAE", "RMSE", "MSE"]
for i in range(0,3):
    liste1_prophet = resultats_prophet["Maladies"][s[i]].tolist()
    liste_prophet = [float(item) for item in liste1_prophet]

    liste1_naive = resultats_naive["Maladies"][s[i]].tolist()
    liste_naive = [float(item) for item in liste1_naive]

    liste1_transformers = resultats_transformers["Maladies"][s[i]].tolist()
    liste_transformers = [float(item) for item in liste1_transformers]
    
    liste1_sarima = resultats_sarima["Maladies"][s[i]].tolist()
    liste_sarima = [float(item) for item in liste1_sarima]
    
    x = np.array([0,1,2,3,4,5,6,7])
    y = np.array(liste_prophet)
    y2 = np.array(liste_naive)
    y4 = np.array(liste_transformers)
    y3 = np.array(liste_sarima)
    all_y = np.transpose(np.array((y, y2,y4,y3)))
    df2 = pd.DataFrame(all_y, columns=['Prophet', 'Naive', 'Transformers','Sarima'])
    df2.plot.bar(figsize=(20, 5));
    
    liste_NAME = resultats_prophet["Maladies"]["DATA_SET_NAME"].tolist()
    my_xticks = liste_NAME
    plt.xticks(x, my_xticks)
    plt.xticks(rotation=45)
    plt.title("Evolution de la " + s[i] + " - MALADIES",fontsize=25)
    plt.xlabel("Noms des DataFrames",fontsize=18)
    plt.ylabel(s[i],fontsize=18)
    plt.legend()
    plt.show()
```

```python
def display_info_df(df, name):
    mean_MAE = []
    mean_MSE = []
    mean_RMSE = []
    median_MAE = []
    median_MSE = []
    median_RMSE = []
    index = []
    for i in df:
        mean_MAE.append(df[i]["MAE"].mean())
        mean_MSE.append(df[i]["MSE"].mean())
        mean_RMSE.append(df[i]["RMSE"].mean())
        median_MAE.append(df[i]["MAE"].median())
        median_MSE.append(df[i]["MSE"].median())
        median_RMSE.append(df[i]["RMSE"].median())
        index.append(i)

    mean = pd.DataFrame({'MAE': mean_MAE, 'MSE': mean_MSE, 'RMSE': mean_RMSE},
                          index = index)
    median = pd.DataFrame({'MAE': median_MAE, 'MSE': median_MSE, 'RMSE': median_RMSE},
                          index = index)
    print("\n\tVALEURS MOYENNES "+ name)
    print(mean)
    
    print("\n\tVALEURS MEDIANES "+ name)
    print(median)

display_info_df(resultats_naive, 'METHODE NAIVE')
display_info_df(resultats_prophet, 'PROPHET')
display_info_df(resultats_sarima, 'SARIMA')
display_info_df(resultats_transformers, 'TRANSFORMERS')
```
